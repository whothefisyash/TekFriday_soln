# Part C:Business Logic Automation

```python
import pandas as pd
```

```python
df=pd.read_csv('datasets/main_loan_base.csv')
df.head()
```

```python
def calculate_risk_score(row):
    try:
        return ((row['missed_repayments'] / 1000) * 2) + (row['loan_amount'] / row['collateral_value']) + (row['interest'] / 2)
    except ZeroDivisionError:
        return None

# Function to classify risk based on score
def classify_risk(score):
    if score is None:
        return "INVALID"
    elif score < 15:
        return "LOW"
    elif 15 <= score <= 25:
        return "MEDIUM"
    else:
        return "HIGH"
```

```python
df['risk_score'] = df.apply(calculate_risk_score, axis=1)
df['risk_level'] = df['risk_score'].apply(classify_risk)
```

```python
# Show sample output
print("Sample EMI Risk Tagging Output:\n")
print(df[['loan_acc_num', 'loan_amount', 'collateral_value', 'missed_repayments', 'interest', 'risk_score', 'risk_level']].sample(10, random_state=42))
```
